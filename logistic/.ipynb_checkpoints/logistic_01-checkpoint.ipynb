1. data1 是40名癌症病人的一些生存资料，
其中，X1表示生活行动能力评分（1~100），X2表示病人的年龄，X3表示由诊断到直入研究时间（月）；
X4表示肿瘤类型，X5把ISO两种疗法（“1”是常规，“0”是试验新疗法）；
Y表示病人生存时间（“0”表示生存时间小于200天，“1”表示生存时间大于或等于200天）

试建立Y关于X1~X5的logistic回归模型 

```python
import numpy as np
import pandas as pd
import sys
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
data = pd.read_table('data1.txt',encoding='gb2312')
x = data.iloc[:,0:6]
y = data.iloc[:,6]
```

数据预处理

```python
#数据预处理
#数据标准化
#划分训练集和测试集
from sklearn import preprocessing 
from sklearn.cross_validation import train_test_split

#划分训练集和测试集
p = 0.2 #设置测试集的比例
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = p)
```

```python
#数据标准化
scaler = preprocessing.StandardScaler().fit(x_train)

#训练集数据标准化
x_train = scaler.transform(x_train)
```

```python
#测试数据标准化
scaler = preprocessing.StandardScaler().fit(x_test)
x_test = scaler.transform(x_test)
```

训练模型

```python
from sklearn.linear_model import LogisticRegression as LR
```

```python
#建立逻辑回归模型
lr = LR()
#用训练数据训练
lr.fit(x_train,y_train)
```

```python
#存储模型
from sklearn.externals import joblib
joblib.dump(lr,'logistic.pkl')
```

```python
#给模型打分
print(u'模型的平均正确率为：%s' % lr.score(x_test, y_test))
```
